```python
def idhm_class(df):
    if df['IDHM']>=0.8:
        return 'Alto Desenvolvimento Humano'
    elif df['IDHM']<0.499:
        return 'Baixo Desenvolvimento Humano'
    else:
        return 'Médio Desenvolvimento Humano'
```

```python
df_idhm['classe'] = df_idhm.apply(idhm_class, axis=1)

df_idhm
```

```python
print(df_idhm.dtypes)
```

```python
#testando a função
print(idhm_class({'idhm': 0.85}))
```

```python
import pandas as pd

# DataFrame simplificado para testar
df_test = pd.DataFrame({
    'idhm': [0.85, 0.65, 0.45]
})

# Aplicar a função no DataFrame simplificado
df_test['classe'] = df_test.apply(idhm_class, axis=1)

print(df_test)
```
