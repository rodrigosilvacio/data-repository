```python
response = requests.get(url)
```

```python
response.text
```

```python
Site = BeautifulSoup(response.text, 'html.parser')
```

```python
Site
```

```python
#dados = Site.find_all('div', class_='base-card base-card--link base-search-card base-search-card--link job-search-card')
```

```python
dados = Site.find_all("div", attrs={"class": "base-search-card__info"})
```

```python
dados
```

```python
armazenando_vaga = []
armazenando_empresa = []
armazenando_localizacao = []
armazenando_tempo = []
armazenando_data = []
armazenando_trat = []
armazenando_link = []

for palavra in dados:
  nome_vaga = palavra.find_all("h3")[0].text.strip()
  armazenando_vaga.append(nome_vaga)

  nome_empresa = palavra.find_all("h4")[0].text.strip()
  armazenando_empresa.append(nome_empresa)

  localizacao = palavra.find_all("span", class_="job-search-card__location")[0].text.strip()
  armazenando_localizacao.append(localizacao)

  tempo_vaga = palavra.find_all("time")[0].text.strip()
  armazenando_tempo.append(tempo_vaga)

  data_vaga = palavra.find_all("time")[0]
  armazenando_data.append(data_vaga)

  data_trat = data_vaga.get('datetime')
  armazenando_trat.append(data_trat)

  try:
    link = palavra.a["href"]
  except:
    link = "Não disponível"

  armazenando_link.append(link)

  print(nome_vaga)
  print(nome_empresa)
  print(localizacao)
  print(tempo_vaga)
  print(data_trat)
  print(link)
  print("-"*50)
```

```python
from datetime import datetime
data_agora = datetime.now()
data_agora
```

```python
base_vagas = {
    "vaga": armazenando_vaga,
    "empresa": armazenando_empresa,
    "localizacao": armazenando_localizacao,
    "tempo": armazenando_tempo,
    "data": armazenando_data,
    "data_trata": armazenando_trat,
    "link" : armazenando_link
}

```

```python
df = pd.DataFrame(base_vagas)
df.to_csv("base_vagas.csv", index=False)
```
