```python
# Filtrando apenas as colunas numéricas
dados_numericos = dados.select_dtypes(include='number')

# Calculando a média das colunas numéricas
medias = dados_numericos.mean()

# Obtendo o índice (nomes das colunas) das médias
colunas_media = medias.index.tolist()

print(colunas_media)
```

```python
# Converter todas as colunas para valores numéricos, forçando a conversão de strings para NaN
dados_numericos = dados.apply(pd.to_numeric, errors='coerce')

# Calcular a média
medias = dados_numericos.mean()

# Obter as colunas das médias
colunas_media = medias.index.tolist()

print(colunas_media)
```

```python
# Remover a coluna 'Unidade da Federação' se ela não for numérica
dados_numericos = dados.drop(columns=['Unidade da Federação'])

# Calcular a média das colunas numéricas
medias = dados_numericos.mean()

# Obter as colunas das médias
colunas_media = medias.index.tolist()

print(colunas_media)
```

```python
# Convert relevant columns to numeric before calculating the mean.
# Assuming the columns you want to calculate the mean for are from index 1 onwards
for column in dados.columns[1:]:
    try:
        dados[column] = pd.to_numeric(dados[column], errors='coerce')  # errors='coerce' will replace invalid values with NaN
    except (ValueError, TypeError):
        print(f"Could not convert column '{column}' to numeric. Check its values.")

# Now calculate the mean for the numeric columns only
# Exclude the "Unidade da Federação" column using iloc[:, 1:]
dados.iloc[:, 1:].mean()
```
